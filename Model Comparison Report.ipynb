# 7. Hyperparameter Tuning
<span style="font-size: 20px;"> 
Hyperparameter tuning was performed to identify the optimal model configuration that minimizes prediction error and improves the model's generalization performance. 

For this task, **RandomizedSearchCV** was used to efficiently explore the hyperparameter space. Compared to Grid Search, Randomized Search evaluates a fixed number of randomly selected parameter combinations, significantly reducing computation time while still producing high-quality results.
</span>
### Evaluation Metric (WMAE)
<span style="font-size: 20px;"> 

The ` Weighted Mean Absolute Error (WMAE)` was selected as the primary evaluation metric because it is the official evaluation metric used in the Walmart Kaggle competition. Holiday weeks have a greater business impact due to increased customer demand; therefore, errors made during these weeks are assigned **5× higher weight** than errors during regular weeks.
<br>
I used direction as `minimize` while creating study **lower WMAE** indicates better forecasting performance. if we use R2 then go for maximize
</span>
### Why WMAE Instead of R²?
<span style="font-size: 20px;"> 
Although R² measures how much variance is explained by the model, it does not differentiate between errors made during holiday and non-holiday periods. In retail demand forecasting, inaccurate predictions during major shopping events can lead to significant inventory and revenue losses. Since WMAE penalizes holiday prediction errors more heavily, it provides a more realistic evaluation of business performance. Therefore, WMAE was chosen for hyperparameter tuning.
</span>

### Hyperparameters Tuned The following XGBoost hyperparameters were optimized:

| Hyperparameter | Purpose |
|---------------|---------|
| n_estimators | No of boosting trees. |
| learning_rate | controls how quickly the model learns. |
| max_depth | Max depth of each DT tree. |
| min_child_weight | controls model complexity as well as reduces overfitting. |
| subsample | % of training samples used for each tree. |
| colsample_bytree | % of features used while building each tree. |
| gamma | Minimum loss reduction required before splitting a node. |

### Hyperparameter Search Summary

- Search Method: **RandomizedSearchCV**
- Number of Trials: **50**
- Cross Validation: **5-Fold Cross Validation**
- Scoring Metric: **Negative Weighted Mean Absolute Error (Negative WMAE)**

After evaluating 50 different parameter combinations, the configuration producing the lowest WMAE was selected as the final model.

#### The optimal model selected by Optuna was `Random Forest` with the following hyperparameters:

## 8.2 Residual Analysis
<span style="font-size: 20px;"> 
Residual analysis was performed to evaluate the prediction errors of the trained model and determine whether any systematic patterns remain after forecasting. A residual is the difference between the actual and predicted weekly sales.
</span>

##### `Residual = Actual Sales − Predicted Sales`


### Distribution of Residuals
- Plotted a **Histogram with KDE (Kernel Density Estimate)** to visualize the distribution of residuals.
- Most residuals are concentrated around **zero**, indicating that the model predicts the majority of weekly sales with relatively small errors.
- A few large positive and negative residuals are present, suggesting that the model struggles to predict some extreme sales values.
- The distribution is **slightly right-skewed**, indicating occasional underestimation during unusually high sales periods.


### Q-Q Plot

- Generated a **Quantile-Quantile (Q-Q) Plot** to compare the residual distribution with a theoretical normal distribution.
- If the residuals are normally distributed, the points should closely follow the reference line.
- Most residuals near the center follow the reference line reasonably well.
- Noticeable deviations occur at both tails, particularly the upper tail, indicating the presence of outliers and extreme prediction errors.
- Therefore, the residuals are **approximately normal near the center but are not perfectly normally distributed**.

# 10. Model Explainability
<span style="font-size: 20px;">
I performed Model Explainability  to understand how the trained model makes its predictions and to ensure that  forecasting results are transparent and interpretable for business stakeholders. 

**LIME (Local Interpretable Model-agnostic Explanations)** was used to explain the contribution of individual features for a specific sales prediction.
</span>

### 10.1 LIME – Local Interpretability

LIME explains a single prediction by approximating the complex machine learning model with a simple local model around the selected instance. This helps identify which features increase or decrease the predicted weekly sales.

### Insights

- The LIME explanation indicates that **Department Average Sales (Dept_Avg_Sales)** had the strongest negative influence on the selected prediction, making it the most influential feature for this instance.
- **Store Size** and **Store Average Sales (Store_Avg_Sales)** also contributed significantly to the prediction, indicating that store characteristics and historical sales strongly influence weekly demand.
- **Week Number** had a positive contribution, suggesting that seasonal trends affected the predicted sales for this observation.
- Features such as **IsHoliday**, **Department**, **Store**, **MarkDown3**, **Economic_Index**, and **Store Type** had comparatively smaller contributions but still influenced the final prediction.
- to summarize , the LIME explanation shows that the model primarily relies on historical sales-related features while incorporating seasonal, promotional, and store-specific factors to generate accurate forecasts.

### Business Interpretation

- Historical sales patterns are the primary drivers of weekly sales predictions.
- Store characteristics and department performance significantly influence demand forecasting.
- Seasonal effects, represented by the week number, contribute positively to sales during certain periods.
- Promotional and economic factors have a supporting influence but are less significant than historical sales information for this prediction.m
